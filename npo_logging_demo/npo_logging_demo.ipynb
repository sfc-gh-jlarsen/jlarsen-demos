# NPO Logging Demo: print() vs logging module

This notebook demonstrates what appears in the event table when a Notebook Project Object (NPO) runs.

**Key Question:** Does `print()` work? Does `logging.info()` work? What shows up in observability?

```python
# ============================================================
# SECTION 1: Setup - Configure Python logging
# ============================================================
import logging
import sys
from datetime import datetime

# Configure root logger - THIS IS REQUIRED for logs to appear in event table
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger('npo_demo')
logger.setLevel(logging.DEBUG)  # Set to DEBUG to test all levels

run_id = datetime.now().strftime('%Y%m%d_%H%M%S')
logger.info(f'=== NPO LOGGING DEMO START (run_id={run_id}) ===')
```

```python
# ============================================================
# SECTION 2: Test print() statements
# ============================================================
# Question: Do print() statements appear in the event table?

print(f'[PRINT-TEST-1] Basic print statement (run_id={run_id})')
print(f'[PRINT-TEST-2] Print with flush=True (run_id={run_id})', flush=True)

# Print to stderr
print(f'[PRINT-TEST-3] Print to stderr (run_id={run_id})', file=sys.stderr)
print(f'[PRINT-TEST-4] Print to stderr with flush (run_id={run_id})', file=sys.stderr, flush=True)

# Print to stdout explicitly
sys.stdout.write(f'[PRINT-TEST-5] sys.stdout.write (run_id={run_id})\n')
sys.stdout.flush()

sys.stderr.write(f'[PRINT-TEST-6] sys.stderr.write (run_id={run_id})\n')
sys.stderr.flush()
```

```python
# ============================================================
# SECTION 3: Test logging module at various levels
# ============================================================
# Question: Which log levels appear in the event table?

logger.debug(f'[LOG-TEST-DEBUG] Debug message (run_id={run_id})')
logger.info(f'[LOG-TEST-INFO] Info message (run_id={run_id})')
logger.warning(f'[LOG-TEST-WARNING] Warning message (run_id={run_id})')
logger.error(f'[LOG-TEST-ERROR] Error message (run_id={run_id})')
logger.critical(f'[LOG-TEST-CRITICAL] Critical message (run_id={run_id})')

# Test with structured data
logger.info(f'[LOG-TEST-STRUCTURED] model_metrics={{"mse": 0.123, "r2": 0.95, "run_id": "{run_id}"}}')
```

```python
# ============================================================
# SECTION 4: Simple ML workflow with logging
# ============================================================
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

logger.info(f'[ML-STEP-1] Generating training data (run_id={run_id})')
np.random.seed(42)
X = np.random.randn(500, 3)
y = 2*X[:, 0] + 3*X[:, 1] - X[:, 2] + np.random.randn(500) * 0.5

logger.info(f'[ML-STEP-2] Training model (run_id={run_id})')
model = LinearRegression()
model.fit(X[:400], y[:400])

logger.info(f'[ML-STEP-3] Evaluating model (run_id={run_id})')
y_pred = model.predict(X[400:])
mse = mean_squared_error(y[400:], y_pred)
r2 = r2_score(y[400:], y_pred)

# Also print the same info for comparison
print(f'[PRINT-ML-RESULT] MSE={mse:.4f}, R2={r2:.4f} (run_id={run_id})')
logger.info(f'[ML-RESULT] MSE={mse:.4f}, R2={r2:.4f} (run_id={run_id})')

logger.info(f'=== NPO LOGGING DEMO COMPLETE (run_id={run_id}) ===')
```
